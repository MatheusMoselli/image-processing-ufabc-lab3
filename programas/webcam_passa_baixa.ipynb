```python
import cv2 as cv
from datetime import datetime
import os

# Item 2

PATH_BASE = "../filtros/exe2_gravacao/"
if not os.path.exists(PATH_BASE):
    os.makedirs(PATH_BASE)

def rodar_filtro(tipo_filtro, tamanho_kernel):
    cap = cv.VideoCapture(0)
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if tipo_filtro == "mediana":
            frame_filtrado = cv.medianBlur(frame, tamanho_kernel)
        elif tipo_filtro == "bilateral":
            frame_filtrado = cv.bilateralFilter(frame, tamanho_kernel, 75, 75)

        cv.imshow(f'Filtro: {tipo_filtro} - Kernel: {tamanho_kernel}', frame_filtrado)
        
        tecla = cv.waitKey(1) & 0xFF
        if tecla == ord('q'):
            break
        elif tecla == ord('s'):
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            nome_arquivo = f"{PATH_BASE}{tamanho_kernel}x{tamanho_kernel}_{tipo_filtro}_{timestamp}.jpg"
            cv.imwrite(nome_arquivo, frame_filtrado)

    cap.release()
    cv.destroyAllWindows()

rodar_filtro("mediana", 3)
rodar_filtro("mediana", 29)
rodar_filtro("bilateral", 3)
rodar_filtro("bilateral", 29)
```
